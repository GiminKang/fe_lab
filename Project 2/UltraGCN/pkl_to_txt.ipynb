```python
f = open('train.txt', 'w')
for i in range(1, len(data['train_indptr'])):
    user = str(i-1)+" "
    for j in range(data['train_indptr'][i-1], data['train_indptr'][i]):
        user += str(data['train_data'][1][j]) + " "
    user = user[:-1]
    f.write(user+"\n")
f.close()
```

```python
f = open('test.txt', 'w')
for i in range(1, len(data['test_indptr'])):
    user = str(i-1)+" "
    for j in range(data['test_indptr'][i-1], data['test_indptr'][i]):
        if data['test_data'][2][j] == 1:
            user += str(data['test_data'][1][j]) + " "
    user = user[:-1]
    f.write(user+"\n")
f.close()
```
